```python
import requests
from bs4 import BeautifulSoup
import csv

# Eingabe- und Ausgabe-CSV-Dateipfade
input_csv_path = 'C:/Users/janlu/Desktop/HTW Studium/HTW Studium Master/3. Semester/W3-BDA Big Data Analytics/EXCEL/Berlins_TOP_100_Dienstleistungen_ueberschrift_korrigiert.csv'
output_csv_path = 'C:/Users/janlu/Desktop/HTW Studium/HTW Studium Master/3. Semester/W3-BDA Big Data Analytics/EXCEL/Berlins_TOP_100_Dienstleistungen_ueberschrift_korrigiert_TEXT.csv'

# Funktion zum Extrahieren der ersten drei Blöcke innerhalb des Haupt-DIVs
def extract_first_three_blocks(soup, main_div_id):
    main_div = soup.find('div', id=main_div_id, class_='servicedetail-view')
    if main_div:
        # Finde alle 'div' Elemente direkt unter dem Haupt-DIV
        blocks = main_div.find_all('div', recursive=False)
        
        # Extrahiere nur die ersten drei Blöcke
        first_three_blocks = blocks[:3]
        extracted_texts = []
        
        for block in first_three_blocks:
            # Text extrahieren und alle \n durch Leerzeichen ersetzen
            text = block.get_text(separator=' ', strip=True).replace('\n', ' ').replace('\r', ' ')
            extracted_texts.append(text)
        
        return extracted_texts
    else:
        return ["Block nicht gefunden", "Block nicht gefunden", "Block nicht gefunden"]

# ID des Hauptinhalts-DIVs
main_content_id = 'layout-grid__area--maincontent'

# CSV-Datei lesen und Daten extrahieren
with open(input_csv_path, newline='', encoding='utf-8') as csvfile:
    csvreader = csv.DictReader(csvfile, delimiter=';')
    results = []

    for row in csvreader:
        name = row['Verwaltungsleistung']
        link = row['Link']
        print(f"Verarbeite: {name} - {link}")

        # HTML-Inhalt von der Webseite abrufen
        try:
            response = requests.get(link)
            response.raise_for_status()  # Überprüfen, ob die Anfrage erfolgreich war
            soup = BeautifulSoup(response.content, 'html.parser')

            # Extrahiere die Texte der ersten drei Blöcke
            block_texts = extract_first_three_blocks(soup, main_content_id)

            # Füge die Ergebnisse zur Liste hinzu
            results.append({
                'Verwaltungsleistung': name,
                'Link': link,
                'Block1': block_texts[0],
                'Block2': block_texts[1],
                'Block3': block_texts[2]
            })

        except requests.exceptions.RequestException as e:
            print(f"Fehler beim Abrufen der Seite {link}: {e}")
            results.append({
                'Verwaltungsleistung': name,
                'Link': link,
                'Block1': "Fehler beim Abrufen",
                'Block2': "Fehler beim Abrufen",
                'Block3': "Fehler beim Abrufen"
            })

# Ergebnisse in eine neue CSV-Datei schreiben
with open(output_csv_path, mode='w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Verwaltungsleistung', 'Link', 'Block1', 'Block2', 'Block3']
    csvwriter = csv.DictWriter(csvfile, fieldnames=fieldnames)

    csvwriter.writeheader()
    csvwriter.writerows(results)

print(f"Die extrahierten Inhalte wurden in {output_csv_path} gespeichert.")
```
